```python
import os
import glob
import pandas as pd
import cv2
```

```python
base_data_dir = '/home/ye/data_sv/'
task = 'keypoint'
img_dir = '/home/ye/data_sv/image/'
column = ['index', 'x', 'y']
data_labels = glob.glob(os.path.join(base_data_dir, task) + '/*.csv')
dataset_dir = 'datasets/MPCT'
```

```python
bbox_width = 40
bbox_height = 40
for label in data_labels:
    video_name = os.path.basename(label).split('catheter_')[1].split('.')[0]
    video_dir = os.path.join(dataset_dir, video_name)
    img_save_dir = os.path.join(video_dir, 'img')
    os.makedirs(img_save_dir, exist_ok=True)
    labels = pd.read_csv(label)
    print(label, video_name, len(labels))
    bbox_annotation = open(os.path.join(video_dir, 'groundtruth_rect.txt'), 'w')
    for i in range(len(labels)):
        idx = labels['index'][i]
        img = cv2.imread(img_dir + video_name + f'_{str(idx)}.jpg')
        if img is not None:
            img_y, img_x, _ = img.shape
            img = cv2.resize(img, (512, 512))
            x = labels['x'][i] * 512 / img_x
            y = labels['y'][i] * 512 / img_y
            row = f'{str(int(x)-bbox_height//2)},{str(int(y)-bbox_height//2)},{bbox_width},{bbox_height}\n'
            img_name = '{:04d}.jpg'.format(i)
            cv2.imwrite(os.path.join(img_save_dir, img_name), img)
            bbox_annotation.write(row)
    bbox_annotation.close()
```

```python
video_list = open('datasets/list/mpct.txt', 'w')
for label in data_labels:
    video_name = os.path.basename(label).split('catheter_')[1].split('.')[0]
    video_dir = os.path.join(dataset_dir, video_name)
    img_save_dir = os.path.join(video_dir, 'img')
    os.makedirs(img_save_dir, exist_ok=True)
    labels = pd.read_csv(label)
    
    
    for i in range(len(labels)):
        idx = labels['index'][i]
        img = cv2.imread(img_dir + video_name + f'_{str(idx)}.jpg')
        if img is not None:
            video_list.write(video_name+'\n)
            break
video_list.close()
```
